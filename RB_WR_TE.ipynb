```python
import nfl_data_py as nfl
import pandas as pd
import json

df = nfl.import_weekly_data([2024])

# RB filter — at least 1 carry or 1 target
rb = df[(df['position'] == 'RB') & ((df['carries'] >= 1) | (df['targets'] >= 1))].copy()

cols = [
    'season', 'week', 'player_id', 'player_name', 'player_display_name',
    'recent_team', 'opponent_team',
    'carries', 'rushing_yards', 'rushing_tds',
    'targets', 'receptions', 'receiving_yards', 'receiving_tds',
    'fantasy_points', 'fantasy_points_ppr'
]
cols = [c for c in cols if c in rb.columns]
rb = rb[cols].copy()

float_cols = rb.select_dtypes(include='float').columns
rb[float_cols] = rb[float_cols].round(2)
rb = rb.sort_values(['player_display_name', 'season', 'week'])
rb = rb.where(pd.notnull(rb), None)

with open('rb_data.json', 'w') as f:
    json.dump(rb.to_dict(orient='records'), f)

print(f"RB rows: {len(rb)}, Unique: {rb['player_display_name'].nunique()}")
```
